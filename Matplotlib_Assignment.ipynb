# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

```python
import numpy as np
#For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
#For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
#For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
#For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

fig.suptitle('Four Line Plots')

plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
company
```

```python
profit
```

"""
A bar plot, also known as a bar chart, is a type of plot used to represent categorical data with rectangular bars. Each bar represents a category, and the height of the bar corresponds to the value or frequency of the category.

Bar plots are commonly used to compare the values of different categories or to track changes over time. They are particularly useful for showing data that can be organized into discrete categories, such as survey responses or product sales by category.

Bar plots are often used in research, business, and other fields to present data in a clear and easy-to-understand way. They can be used to highlight trends, identify patterns, and communicate insights to a wide range of audiences.

In Matplotlib, bar plots can be created using the bar() function or the barh() function, depending on whether you want the bars to be vertical or horizontal. Additional parameters can be used to customize the appearance of the bars, such as color, width, and alignment.
"""

```python
plt.figure(figsize=(6,4))
plt.bar(company,profit,color="#0F1A72")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Profit bar of MNCs")
plt.grid()
```

```python
plt.figure(figsize=(6,4))
plt.barh(company,profit,color="#0F1A72")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Profit bar of MNCs")
plt.grid()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
box1
```

```python
box2
```

"""
A box plot, also known as a box-and-whisker plot, is a type of plot used to represent the distribution of a dataset. It shows the median, quartiles, and outliers of the data in a compact way.

The box in the box plot represents the interquartile range (IQR), which is the distance between the 25th and 75th percentiles of the data. The median is represented by a line in the box. The "whiskers" extend from the box to show the range of the data, excluding outliers, which are represented as individual points.

Box plots are often used to compare the distribution of two or more datasets, to identify outliers or anomalies, and to visualize the spread and skewness of the data.
"""

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot of Two Datasets')

plt.show()
```
