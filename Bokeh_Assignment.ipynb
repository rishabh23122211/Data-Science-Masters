```python
# Q1. How can you create a Bokeh plot using Python code?
```

```python
## Bokeh is a Python library for creating interactive data visualizations in web browsers. To create a Bokeh plot using Python code, you can follow these steps:
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, output_file, show


x_data = [1, 2, 3, 4, 5]
y_data = [6, 7, 2, 4, 5]

p = figure(title="My Plot", x_axis_label="X Label", y_axis_label="Y Label")

p.circle(x_data, y_data, size=10, color='navy')

output_file("my_plot.html")
show(p)
```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
## In Bokeh, glyphs are visual shapes that represent data points in a plot, such as circles, squares, triangles, and lines. Glyphs are added to a plot using the glyph method of the figure object.
```

```python
from bokeh.plotting import figure, show
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="My Plot", x_axis_label="X Label", y_axis_label="Y Label")
p.line(x, y, line_width=2, line_color='blue')
p.circle(x, y, size=5, fill_color='white', line_color='blue')
p.square(x, y, size=5, fill_color='white', line_color='red')

show(p)
```

```python
# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?
```

Bokeh is a Python library for creating interactive visualizations, including plots, charts, and dashboards. Bokeh's server allows you to create web applications that can update data and visualizations in real-time, without needing to reload the entire page.

Bokeh server operates by creating a persistent connection between the web browser and a running Python process. This connection enables the browser to receive updates from the server in real-time and send user interactions back to the server for processing.

To create an interactive plot using Bokeh server, you first create a Bokeh plot as you would normally do using the Bokeh library. Then, you define a callback function that will update the plot based on user input or new data. The callback function is registered with the Bokeh server using the on_change method of the data source object.

Once the callback function is registered, the Bokeh server will handle all communication between the browser and the Python process, ensuring that updates to the plot are sent to the browser in real-time, without needing to reload the page.

To launch a Bokeh server application, you can use the bokeh serve command from the command line, followed by the name of the Python script containing your application code. The server will then start listening for incoming connections on a specified port, and you can access the application in your web browser by navigating to the appropriate URL.
