```python
plt.scatter(x,y)
plt.xlabel("This is my x- axis")
plt.ylabel("This is my y-axis")
plt.title("This is my x vs y graph")
```

```python
plt.scatter(x,y,c="g")
plt.xlabel("This is my x- axis")
plt.ylabel("This is my y-axis")
plt.title("This is my x vs y graph")
```

```python
plt.scatter(x,y,c="#C133FF",alpha = 0.3)
plt.xlabel("This is my x- axis")
plt.ylabel("This is my y-axis")
plt.title("This is my x vs y graph")
```

```python
plt.scatter(x,y,c="#C133FF",marker="*")
plt.xlabel("This is my x- axis")
plt.ylabel("This is my y-axis")
plt.title("This is my x vs y graph")
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,c="#C133FF",alpha=0.4,marker="*")
plt.xlabel("This is my x- axis")
plt.ylabel("This is my y-axis")
plt.title("This is my x vs y graph")
plt.grid()
```

```python
# 3d_plot
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
z
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection="3d")
ax.scatter(x,y,z,color="#72430F",alpha=0.9, marker="*")
plt.xlabel(" x- axis")
plt.ylabel("y-axis")
plt.title("3d- graph")
plt.show()
```
